```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Generar conjuntos de datos
n_samples = 1500

# Configuración inicial
np.random.seed(0)

# Conjuntos de datos
datos_circulos_ruidosos = datasets.make_circles(n_samples=n_samples, factor=.5, noise=.05)
datos_lunas_ruidosas = datasets.make_moons(n_samples=n_samples, noise=.05)
datos_blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)

# Datos distribuidos anisotrópicamente
X, y = datasets.make_blobs(n_samples=n_samples, random_state=170)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
aniso = (X_aniso, y)
```

```python
# Graficar datos de círculos ruidosos
X_circulos, _ = datos_circulos_ruidosos
plt.figure(figsize=(8, 6))
plt.scatter(X_circulos[:, 0], X_circulos[:, 1], c='black', s=10)
plt.show()

# Graficar datos de lunas ruidosas
X_lunas, _ = datos_lunas_ruidosas
plt.figure(figsize=(8, 6))
plt.scatter(X_lunas[:, 0], X_lunas[:, 1], c='black', s=10)
plt.show()

# Graficar datos de blobs
X_blobs, _ = datos_blobs
plt.figure(figsize=(8, 6))
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c='black', s=10)
plt.show()

# Graficar datos anisotrópicos
X_sin_estructura, _ = aniso
plt.figure(figsize=(8, 6))
plt.scatter(X_sin_estructura[:, 0], X_sin_estructura[:, 1], c='black', s=10)
plt.show()
```

```python
# Convertir cada conjunto de datos a un DataFrame
df_circulos = pd.DataFrame(X_circulos, columns=['Caracteristica 1', 'Caracteristica 2'])

df_lunas = pd.DataFrame(X_lunas, columns=['Caracteristica 1', 'Caracteristica 2'])

df_blobs = pd.DataFrame(X_blobs, columns=['Caracteristica 1', 'Caracteristica 2'])

df_aniso = pd.DataFrame(X_aniso, columns=['Caracteristica 1', 'Caracteristica 2'])

# Mostrar los primeros registros de cada DataFrame para verificar
print("DataFrame de Círculos Ruidosos:")
print(df_circulos.head())

print("\nDataFrame de Lunas Ruidosas:")
print(df_lunas.head())

print("\nDataFrame de Blobs:")
print(df_blobs.head())

print("\nDataFrame de Datos Anisotrópicos:")
print(df_aniso.head())
```
